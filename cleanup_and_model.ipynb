# Data Clean-up

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('recruitmentdataset-2022-1.3.csv')      
```

```python
data = data.drop(['Id', 'gender', 'age', 'nationality', 'company'], axis = 1)

data.columns = ['sport', 'university_grade', 'debateclub', 'programming_exp', 'international_exp', 
                'entrepreuner_exp', 'languages', 'exact_study', 'degree', 'decision']

sport_mapping = {'Swimming': 0
                 ,'Golf': 1
                 ,'Running': 2
                 ,'Cricket': 3
                 ,'Chess': 4
                 ,'Tennis': 5
                 ,'Football': 6
                 ,'Rugby': 7
                }
degree_mapping = {'bachelor': 0
                  ,'master': 1
                  ,'phd': 2
                 }
data = data.assign(sport = data.sport.map(sport_mapping))
data = data.assign(degree = data.degree.map(degree_mapping))
data
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

X_train
```

```python
rf_clf = RandomForestClassifier(n_estimators = 7, criterion = 'gini')   
rf_clf.fit(X_train,y_train)
```

```python
parameters = {'n_estimators': [5, 6, 7, 8, 9],
             'criterion': ['gini', 'entropy'],
             'max_features': ['auto', 'sqrt', 'log2']}
clf = GridSearchCV(RandomForestClassifier(), parameters)
clf.fit(X_train, y_train)
clf.best_params_
```

```python
y_predict = rf_clf.predict(X_test)
accuracy_score(y_test,y_predict)
```
