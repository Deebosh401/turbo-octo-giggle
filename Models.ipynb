## Model 1 (three parameters)

```python
ind_model1b = indicators_a.drop(['university_grade', 'programming_exp', 'international_exp', 'exact_study', 'entrepreuner_exp', 'company'], axis = 1)
X_model1b = ind_model1b.iloc[:, :-1]
y_model1b = ind_model1b.iloc[:, -1]

max_depth=[1, 2, 3]
n_estimators = [100, 150, 200]
param_grid = dict(max_depth=max_depth, n_estimators=n_estimators)

# Define list of scores
scores = ['accuracy', 'f1']

# Build the grid search
model1b = RandomForestClassifier(n_estimators = n_estimators, max_depth = max_depth, random_state = 100, n_jobs = -1)
grid1b = GridSearchCV(estimator = model1b, param_grid = param_grid, cv = 5, scoring = scores, refit = 'accuracy', n_jobs = -1)
validation1b = grid1b.fit(X_model1b, y_model1b)
```

```python
val1b_results = pd.DataFrame(validation1b.cv_results_)
val1b_results = val1b_results.drop(['mean_fit_time', 'std_fit_time', 'mean_score_time', 'std_score_time', 'params'], axis = 1)
print("Best hyperparameters for Model 1: ", validation1b.best_params_)
print("Mean accuracy over 5 folds for best hyperparameters: ", val1b_results['mean_test_accuracy'].max())
val1b_results = val1b_results.sort_values(by=['rank_test_accuracy'], ascending=True)
val1b_results
```
